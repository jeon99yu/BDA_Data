## <데이터 병합>
- 두 개 이상 데이터프레임을 병합하는 과정
    - 병합이라는 것은 데이터프레임을 합치는 개념
    - 왜 합칠까?
        - 내가 원하는 분석을 하기 위해서는 두 개의 테이블을 합쳐야 하는 경우가 발생할 수 있고 ML 에서 train,test 등으로 데이터를 나눠야 하는 경우 등 요구사항에 따라 데이터를 나누거나 병합해야 하는 경우가 발생할 것
        
    - 병합은 결국 합치는 개념   
    
    - **merge()**
        - 두 개의 테이블을 합칠 때 공통된 기준 컬럼을 잡고 병합한다.
        - 교집합에 개념으로 공통된 컬럼을 하나 잡고 데이터를 병합한다.
        - 병합하는 조건 inner (교집합), left , right, outer join 방법이 있다.
        
    - **concat()**
        - 두 데이터프레임을 공통된 기준 없이 그냥 df1 + df2 를 합치게 된다.
        - 두 가지로 생각할 수 있다.
            - 컬럼 기준으로 합치는 경우 (열기준)
            - 인덱스 기준으로 합치는 경우 (행기준)
            
    - join도 있음

```python
import pandas as pd
import seaborn as sns
```

```python
test1=pd.DataFrame({'학회원번호':[1,2,3,4,5],
             '학회원분반':['python','sql','r','데이터분석입문ml','데이터분석입문통계']})

test2=pd.DataFrame({'학회원번호':[1,2,3,4,5],
             '상벌점현황':[10,20,30,40,50]})

display(test1, test2)
```

merge()로 병합을 한다면?

```python
#공동된 기준 컬럼은? => '학회원번호'
pd.merge(test1,test2)
```

-> 먼저 사용한 데이터프레임 기준이 된다.

```python
pd.merge(test2,test1)
```

## 데이터 병합시 유의해야 하는 점

- 항상 공동된 컬럼의 공동된 값이 존재하진 않을 수 있다.
- 즉, 모두 1,2,3,4,5 두 테이블이 다 같진 않을 수 있다.

```python
test_a=pd.DataFrame({'학회원번호':[1,2,3,4,5],
             '학회원분반':['python','sql','r','데이터분석입문ml','데이터분석입문통계']})


test_b= pd.DataFrame({'학회원번호':[1,2,3,8,7],
             '상벌점현황':[10,20,30,40,50]})
```

```python
display(test_a, test_b)
```

## how = 'inner' 교집합의 개념
- how의 다양한 방법에 따라서 데이터의 출력이 달라진다.
    - left, right, outer

```python
pd.merge(test_a,test_b, how='inner') 
```

```python
pd.merge(test_a,test_b, how='left') # left(=test_a)가 기준
```

-> test_a는 다 출력, test_b는 잘릴수도 있음!

```python
display(pd.merge(test_b,test_a, how='right'),pd.merge(test_a,test_b, how='right'))
```

```python
pd.merge(test_a,test_b, how='outer')
```

## concat 
- df, df를 덩어리로 합치게 된다.
- pd.concat([df1, df2], axis=0,1)

```python
pd.concat([test1,test2])
```

```python
pd.concat([test1,test2],axis=0) #행(row)을 기준으로, 위아래로 결합
```

```python
pd.concat([test1,test2],axis=1) #열(column)을 기준으로, 옆으로 결합
```

### groupby 문법
- 데이터 분석시 우리가 요약하거나, 집계해서 비교하거나, 잘라서 부분적인 비교 분석할 경우 
- groupby 그룹을 묶어서 집계하는 것
- groupby 진행 시 
    - 기준컬럼(집계할 기준), 요약통계치 (집계할 함수)
    -  pclass 기준컬럼으로 survived 값을 mean 평균으로 살펴봄
    
- groupby() 데이터를 묶어서 집계함수를 통해서 출력한다.
    - 기본적인 원리를 이해하고 있어야 한다.

```python
df = sns.load_dataset('titanic')
```

```python
#데이터가 pclass로 묶인 상태다.
pcl_grp =df.groupby('pclass')
```

```python
# 그룹으로 묶인 걸 데이터로 확인하려면
# pclass 1,2,3 
for i,j in pcl_grp:
    print(i,'class로 묶인 그룹')
    print(j)
```

```python
# 그룹별로 데이터를 추출할 수 있다.
pcl_grp.get_group(1)
```

```python
pcl_grp.get_group(3)
```

-> pclass만 묶었지만, pclass, sex 같이 묶을 수 있다. 

```python
pcl_grp2 =df.groupby(['pclass','sex'])
pcl_grp2
```

```python
# 그룹으로 묶인 걸 데이터로 확인하려면
# pclass 1,2,3 
for i,j in pcl_grp2:
    print(i,'묶인 그룹')
    print(j)
```

```python
pcl_grp2.get_group((2, 'female'))[['age','fare']].mean()
```

```python
pcl_grp2[['age','fare']].mean()
```

## 집계 요약함수
- agg함수를 사용해서 내가 원하는 요약통계치를 볼 수 있다.
- group_by한 상태에서.agg(내가 볼 요약통계치), 2개 이상이면 리스트로 담아주세요!
- agg함수는 내가 만든 함수도 넣어서 사용할 수 있다.

```python
pcl_grp2[['age','fare']].agg(['mean','max','median'])
```

```python
def min_max(x):
    return x.max()- x.min()
```

```python
pcl_grp2[['age','fare']].agg(min_max)
```
