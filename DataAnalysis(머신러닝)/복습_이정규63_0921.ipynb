### 상황에 따른 판다스 기초 문법
- 1. 데이터 변수명을 바꿔야 하는 경우

```python
## 예시 데이터를 만들어 보자!
## 2차원 구조라 행과 열, 즉 컬럼을 잡아서 값을 매핑하는 형태로 데이터프레임을 만들어야 한다!

df=pd.DataFrame({'분반':['파문기','SQL','모델링','데이터분석'],
             '인워ㄴ':[100,200,300,400]})
```

```python
## 변수명을 바꿀 때 rename
## 데이터프레임.rename(columns = {'기존컬럼':'변경컬럼'})
df.rename(columns = {'인워ㄴ':'인원'})
```

- 데이터프레임은 원본데이터를 보존하려는 성질이 있다보니 바로 바뀌지 않는다.
- 원본 데이터를 바꾸려면 3가지를 생각해야 한다.

- 1. 원본데이터 변수명에 그대로 원본데이터를 넣으면 된다. 변수 사용해서 변수교체
- 2. 바로 문법에서 inplace=True 대치를 바로 진행하는 경우
- 3. copy() 원본데이터를 카피하여서 실제 원본데이터는 그대로 두고, 내가 분석할 데이터만 변경하여 사용한다.

- why?
     - 어느 순간 데이터를 분석하다보면 뭐가 원본이고 무엇을 하고 있는지? 가끔 혼선이 올 경우
     - A로 전처리 했는데- B로 전처리 해야 한다. 원본 없다? 귀찮은 일
     - 이런 여러 예외 케이스를 대응하기 위해서는 원본 데이터는 그대로 두는 게 처음부터 습관! 좋다!

```python
df_1=df.copy()
df_2=df.copy()
```

```python
#변수 교체
df_1=df_1.rename(columns = {'인워ㄴ':'인원'})
```

```python
#이 문법을 사용하면서 바로 원본을 수정하겠다 선언하는 게 inplace True
df_2.rename(columns = {'인워ㄴ':'인원'}, inplace=True)
```

- 2. 파생변수 만들기
    - 기존의 변수들을 변형하여 새로운 변수를 만드는 것

```python
bda_class=pd.DataFrame({'이름':['김길동','이길동','박길동','최길동','홍길동'],
        '상점':[10,20,30,40,50],
             '벌점':[1,2,3,5,8]})
```

```python
## 파생변수는 기존의 변수를 이용해서 새로운 편수를 만드는 것
## 상점 - 벌점 = 최종상태의 점수
## 시리즈형태로 계산하면 빠르게 연산이 가능하다.
bda_class['최종상태']=bda_class['상점']-bda_class['벌점']

## 파생변수를 만드는 게 좋은가?
## 파생변수는 기존 변수의 특성을 가지고 만든 것
## 변수라는 게 추가 된다는 것-> 차원이 추가 된다는 것
## 차원이 추가된다는 것은 -> 정보가 추가되는 것, 복잡해 진다.
## 데이터 타입에 따라서 파생변수를 바라보는 것도 달라져야 한다.
## 파생변수는 정보를 더 알수도 있지만 이게 꼭 필요한가?
```

```python
#최종상태에 따른 pass, fail 컬럼까지 추가해서 만들어 보자!
#기준이 필요하다. 최종상태가 20점 이상이면 pass 20점 미만이면 Fail이다.
bda_class
```

```python
for i in bda_class['최종상태']:
    c=[]
    if i >=20:
        c.append('pass')
    else:
        c.append('fail')
        
#    생각보다 쉽지 않은 for, if문 전처리!
```

```python
## numpy와 함께 응용해서 사용하면 쉽게 데이터프레임 문법을 사용할 수 있다.
## apply 함수와 사용해서 한 번에 데이터프레임을 문법 적용시킬 수 있다.

## numpy 간단하게 전처리할 예정
## numpy.where 문법으로 전처리할 데이터를 서치할 예정 
import numpy as np
#위치 찾아주는 문법
#np.where(데이터프레임의 특정 조건을 넣고 그거에 따른 컬럼의 값을 대입할 수 있다.)

bda_class['합불결과']=np.where(bda_class['최종상태']>=20,'pass','fail')
```

```python
## 기존에 있던 데이터는 대부분 수치형 데이터 
## 합불결과 -> 문자형, 범주형 개념
## 범주형 데이터를 빠르게 볼 수 있는 문법
## value_counts()
bda_class['합불결과'].value_counts()
```

- 데이터의 추출을 위한 두 가지 관점
    - 열 기준
    - 행 기준
        - loc -> Location (인덱스) - 처음 생성된 데이터프레임의 인덱스
        - iloc -> Integer Location (넘버링된 인덱스)  - 현재 기준의 데이터프레임의 순서 인덱스
        - loc[인덱스]
        - iloc[인덱스]
        - 안 되는 예외 조건들이 있다.
        
- loc, iloc을 기준으로 행과 열 모두 접근할 수 있다.
    - 파이썬은 기본적으로 행,열 순서로 만들어진다.
        - loc 컬럼명을 명시적으로 적어줘야 한다.
        - iloc 뒤에 컬럼도 모두 다 숫자 int 만들어야 한다.
