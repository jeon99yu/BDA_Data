## 필수과제1
- q1 - pclass 3 이고, age 20세 미만인 남성과, 여성의 카운트 수와 생존수를 비교해 주세요.

```python
# 3등석 승객 중 20세 이하 남성과 여성 필터링
under_20_class_3 = tt[(tt['pclass'] == 3) & (tt['age'] < 20)]
```

```python
# 성별에 따른 총 인원 수와 생존자 수 비교
comparison = under_20_class_3.groupby('sex').agg(
    total_count=('sex', 'count'),
    survived_count=('survived', 'sum')).reset_index()
```

```python
comparison
```

- 결과분석: 남자 총인원은 더 많았지만 여자가 남자보다 생존수가 더 많았다

## 필수과제2
- 요금에 따라 3가지 등급으로 나눠주시고 (요금의 등급의 기준은 기초통계치 분위수 값으로 기준)
- 해당 분위수에 따라 남성과 여성을 추출하고 카운팅 수, 생존수 비교해 주세요

```python
tt
```

```python
# 1. 결측치 제거 
tt = tt.dropna(subset=['fare'])
```

```python
# 2. 가격(fare)을 3개의 등급으로 나누기 (분위수 기반)
tt['Fare_Quantile'] = pd.qcut(tt['fare'], 3, labels=['Low', 'Medium', 'High'])
```

```python
# 3. 성별에 따른 총 인원 수와 생존자 수 비교 (가격 분위수별)
comparison2 = tt.groupby(['Fare_Quantile', 'sex']).agg(
    total_count=('sex', 'count'),
    survived_count=('survived', 'sum')).reset_index()
```

```python
comparison2
```

- 요금이 높은 이용객일수록 생존자가 많았고 주로 남자보다는 여자가 생존자가 더 많았다

## 필수과제3
- 혼자온 사람과 혼자 오지 않은사람들 중에서 둘을 비교하는데 공통 조건은 혼자 온 사람의 성별 여부에 따른 전체 카운팅과 생존수 비교해 주세요!

```python
tt
```

```python
# 1. 결측치 처리
tt = tt.dropna(subset=['sibsp', 'parch', 'sex', 'survived'])
```

```python
# 2. '혼자 온 승객' 열 추가
# SibSp(형제자매/배우자)와 Parch(부모/자녀)가 모두 0인 경우 혼자 온 것으로 간주
tt['Came_Alone'] = (tt['sibsp'] == 0) & (tt['parch'] == 0)
```

```python
# 3. 혼자 온 승객의 성별에 따른 총 인원 수와 생존자 수 비교
comparison3 = tt[tt['Came_Alone'] == True].groupby('sex').agg(
    total_count=('sex', 'count'),
    survived_count=('survived', 'sum')).reset_index()
```

```python
comparison3
```

- 혼자 온 승객은 남자가 더 많았고 생존률은 혼자 온 여자가 월등히 높았다
