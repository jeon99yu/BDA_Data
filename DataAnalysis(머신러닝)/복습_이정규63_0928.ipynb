# Query()
- 쿼리 -> 내가 원하는 데이터의 조건을 질의한다.
- 그 조건을 명확하게 어떤 식으로 작성해야 하는지? 문법적인 부분을 이해하고 데이터를 추출하자!
- 데이터프레임의 문법
- 데이터프레임.query('조건')

- 데이터프레임에 작동하는 문법

```python
tt.info()
```

```python
# pclass 3인 데이터만 추출
# tt.query('pclass == 3') - int64 수치형 데이터의 조건
tt.query('pclass == 3')
```

```python
## 문자열 데이터 추출
# tt.query('sex == male') - object 문자형 데이터의 조건
# tt.query("sex == 'male'") - object 문자형 데이터의 조건
tt.query('sex == male')
```

```python
#tt.query(‘Sex == “male”’) 
#python 기초 문법
#문자열 값을 '' 문자열로 인식하게 만들어야 한다!
tt.query("sex == 'male'") 
```

```python
## 조건이 여러 개인 경우!
## 연산자가 필요하다, 비교, 논리 등등
## 여러 연산자를 사용해서 다중조건을 만들자!
## and, or, not 

## 나이가 20살 이상 40살 미만 데이터

tt.query('age >=20 & age <40 ')
```

```python
## 나이가 20살 이상 40살 미만의 여성의 데이터
tt.query('''age >=20 & age <40 and sex =='female' ''')
```

```python
tt.query('''age >=20 & age <40 or sex =='female' ''')
```

```python
# and, or not, 연산 순서 다르다.
## 데이터프레임의 장점은 데이터를 재구조화 할 수 있다.
##### 결국 끝이 데이터프레임이면-> 다시 데이터프레임처럼 문법을 사용할 수 있다.

#tt.query('''age >=20 & age <40 or sex =='female' ''')

tt.query('''age >=20 & age <40 or sex =='female' ''').query('age <30')
```

```python
#데이터를 재구조화 할 수 있다.
# 복습하시면서 질문을 하고 그 질문을 찾아가는 과정-> 데이터재구조화 
tt.query('''age >=20 & age <40 and sex =='female' ''').iloc[0:300][['survived','sex','pclass']].query('survived == 0')
```

- df[[불린값]] - 불린값 형태의 데이터 추출
- df[[조건에 따른 참값]] - 해당 조건으로 데이터가 추출이 된다.
- df[[참인 조건]] - 참인 조건으로 데이터가 추출
- 참인 조건 시리즈 형태로 데이터를 넣을 수 있다.
- 다중 조건은 어떤 식으로 할까?
- tt[(tt['sex']=='male') and,or 등(다른조건)...n조건]

```python
tt[tt['sex']=='female']
```

```python
# 다중조건
tt[(tt['sex']=='female') & (tt['pclass']==3)]
```

```python
#먼저 불린값으로 데이터를 추출 후 -> 인덱스 loc 1:300 값을 추출
tt[(tt['sex']=='male') & (tt['pclass']==3)].loc[1:300]
```

```python
#뒤의를 iloc integer location 인덱스 값이 다르다.
tt[(tt['sex']=='male') & (tt['pclass']==3)].iloc[1:300]
```

```python
## 어떤 조건을 먼저 사용해야 합니까? 내가 데이터를 찾을 때
tt.loc[1:300][(tt['sex']=='male') & (tt['pclass']==3)]
```

```python
tt[(tt['sex']=='male') & (tt['pclass']==3)].loc[1:300]
```

- 타이타닉 데이터의 남성과 여성 데이터를 추출하고 
- 남성도 여성도 pclass == 3 , age 20세 미만의 생존율

- 데이터는 남성과, 여성 두 개의 테이블이 필요하다.
- pclass ==3 둘다 age 20세 미만
- 생존율 

```python
## 공통된 조건으로 먼저 날려버리자!
## pclass ==3 둘다 age 20세 미만
#tt[(tt['age'] <20) & (tt['pclass']==3)]

## 남성과 여성에 대한 비교
## 생존율추출
tt[(tt['age'] <20) & (tt['pclass']==3)][['sex','survived']][tt[(tt['age'] <20) & (tt['pclass']==3)][['sex','survived']]['sex']=='male']
```

```python
tt[(tt['age'] <20) & (tt['pclass']==3)][['sex','survived']][tt[(tt['age'] <20) & (tt['pclass']==3)][['sex','survived']]['sex']=='female']
```
