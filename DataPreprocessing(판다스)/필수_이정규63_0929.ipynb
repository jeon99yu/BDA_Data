```python
# 1. model_year 기준으로 그룹화하여 각 연도별로 가장 높은 연비(mpg)를 가진 차량 찾기
group_by_year = mpg.groupby('model_year') 
group_by_year
```

```python
# 2. 각 그룹에서 연비(mpg)가 가장 큰 값 찾기
max_mpg_per_year = group_by_year['mpg'].max()  # 연도별 최대 mpg 추출
max_mpg_per_year
```

```python
# 3. 가장 높은 연비를 가진 차량의 이름을 찾기 위해 merge
result = pd.merge(mpg, max_mpg_per_year, on=['model_year', 'mpg'])
result
```

```python
result[['origin', 'name', 'weight','cylinders', 'mpg']]
```

```python
sorted_result = result[['origin', 'name', 'weight','cylinders', 'mpg']].sort_values(by='mpg', ascending=False)
sorted_result
```
