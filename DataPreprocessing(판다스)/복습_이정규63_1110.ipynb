## `IterativeImputer`의 주요 하이퍼파라미터

1. **estimator**: 누락된 값을 추정하기 위해 사용할 추정기입니다. 기본값은 `BayesianRidge()`이며, `LinearRegression`, `DecisionTreeRegressor`, `KNeighborsRegressor` 등 다른 회귀 추정기로 변경할 수 있습니다.

2. **max_iter**: 최대 반복 횟수로, 누락된 값을 추정하는 최대 사이클 수입니다. 기본값은 `10`입니다.

3. **random_state**: 난수 생성기의 시드 값을 설정합니다. 동일한 결과를 재현하고 싶을 때 유용합니다.

4. **initial_strategy**: 누락된 값을 초기에 대치할 전략으로, 'mean', 'median', 'most_frequent', 또는 'constant' 중 하나를 선택할 수 있습니다. 기본값은 'mean'입니다.

5. **imputation_order**: 특성을 대치하는 순서를 결정합니다. 'ascending', 'descending', 'roman', 'arabic', 또는 'random' 중에서 선택할 수 있습니다. 기본값은 'ascending'입니다.

6. **skip_complete**: True로 설정하면, 완전한 케이스(누락된 값이 없는 경우)를 계산에서 제외합니다. 기본값은 False입니다.

7. **add_indicator**: 누락된 특성에 대한 추가적인 지표(누락된 값이 있음을 나타내는 이진 변수)를 추가할지 여부를 결정합니다. 기본값은 False입니다.

8. **tol**: 반복을 멈출 임계값으로, 이전과 현재 반복 사이의 평균 제곱 차이가 이 값보다 작을 경우 반복이 중지됩니다. 기본값은 `1e-3`입니다.

9. **n_nearest_features**: 각 특성을 대치할 때 고려할 다른 특성의 수입니다. None으로 설정하면 모든 특성이 사용됩니다. 기본값은 None입니다.

10. **verbose**: 실행 중 출력 메시지의 양을 제어합니다. 값이 클수록 더 많은 메시지가 출력됩니다. 기본값은 `0`입니다.

11. **sample_posterior**: True로 설정하면, BayesianRidge 추정기를 사용할 때 샘플링을 통해 후방 분포에서 값을 추정합니다. 기본값은 False입니다.

## `SimpleImputer` 하이퍼파라미터:

1. **missing_values**: 누락된 값을 나타내는 placeholder입니다. 기본값은 `np.nan`입니다. 숫자나 NaN 외에도 다른 표현을 사용할 수 있습니다.

2. **strategy**: 누락된 값을 대체할 방법을 결정합니다. 가능한 값은 "mean", "median", "most_frequent", "constant"입니다. 기본값은 "mean"입니다.

3. **fill_value**: `strategy="constant"`를 사용할 때 누락된 값을 대체하기 위한 상수 값을 설정합니다. 기본값은 `None`입니다. 이 값을 설정하면 모든 누락된 값을 해당 값으로 대체합니다.

4. **copy**: `True`인 경우, 원본 데이터는 변경되지 않고 새 데이터 배열이 생성됩니다. `False`일 경우, 원본 데이터가 대체됩니다. 기본값은 `True`입니다.

5. **add_indicator**: `True`인 경우, 누락된 값이 있는 특성에 대한 인디케이터(지시자) 변수가 결과에 추가됩니다. 기본값은 `False`입니다.

## `KNNImputer` 하이퍼파라미터:

1. **missing_values**: 누락된 값을 나타내는 placeholder입니다. `SimpleImputer`와 마찬가지로 기본값은 `np.nan`입니다.

2. **n_neighbors**: 누락된 값을 대체하는 데 사용할 가장 가까운 이웃의 수입니다. 기본값은 5입니다.

3. **weights**: 이웃을 가중치하는 방법입니다. 가능한 값은 "uniform", "distance" 또는 사용자 정의 함수입니다. "uniform"은 모든 이웃에게 동일한 가중치를, "distance"는 가까운 이웃에 더 높은 가중치를 부여합니다. 기본값은 "uniform"입니다.

4. **metric**: 거리 계산에 사용할 메트릭입니다. 기본값은 "nan_euclidean"으로, 유클리드 거리를 계산할 때 누락된 값이 있는 경우를 처리합니다.

5. **copy**: `True`인 경우, 원본 데이터는 변경되지 않고 새 데이터 배열이 생성됩니다. `False`일 경우, 원본 데이터가 변경됩니다. 기본값은 `True`입니다.

6. **add_indicator**: `True`인 경우, 누락된 값이 있는 특성에 대한 인디케이터(지시자) 변수가 결과에 추가됩니다. 기본값은 `False`입니다.

## `interpolate()` 메소드

### 1. `linear`
- 기본값입니다.
- 결측치를 양쪽의 값으로부터 선형적으로 추정합니다.
- 시계열 데이터에서 일반적으로 사용되며, 데이터가 비교적 균일하게 분포할 때 적합합니다.

### 2. `time`
- 시간 간격을 고려하여 선형적으로 보간합니다.
- 시계열 데이터의 인덱스가 `DatetimeIndex`일 때 유용합니다.

### 3. `index`, `values`
- 인덱스 값에 기반하여 선형적으로 보간합니다.
- `index`는 인덱스 자체를, `values`는 인덱스의 값들을 기준으로 보간합니다.

### 4. `pad`, `ffill`
- 앞 방향으로 이전의 유효한 값으로 결측치를 채웁니다.

### 5. `bfill`, `backfill`
- 뒷 방향으로 다음의 유효한 값으로 결측치를 채웁니다.

### 6. `nearest`
- 결측치에 가장 가까운 인덱스의 값으로 채웁니다.

### 7. `zero`, `slinear`, `quadratic`, `cubic`, `spline`
- `spline`은 차수(order)를 지정할 수 있는데, 예를 들어 `order=2`는 이차 스플라인, `order=3`은 삼차 스플라인 보간을 의미합니다.
- `zero`는 0차 스플라인(계단 함수), `slinear`는 선형, `quadratic`은 이차, `cubic`은 삼차 스플라인 보간입니다.

### 8. `polynomial`
- 다항식 보간을 수행합니다. `order` 옵션으로 다항식의 차수를 지정해야 합니다.

### 9. `piecewise_polynomial`
- 각 세그먼트에서 다항식 보간을 수행하는 조각별 다항식 보간법입니다.

### 10. `from_derivatives`
- 미분 값으로부터 데이터를 보간합니다. 이 방법은 고급 사용자를 위한 것으로, 데이터의 미분값을 알고 있을 때 사용할 수 있습니다.

### 11. `akima`
- Akima의 3차 스플라인 보간법을 사용합니다. 데이터 포인트 사이를 매끄럽게 연결하며, 국부적인 데이터 변동성에 민감하게 반응합니다.

### 12. `krogh`, `piecewise_polynomial`, `pchip`, `akima`, `cubicspline`
- 각각 다른 종류의 스플라인 또는 다항식 기반 보간 방법입니다. 데이터의 특성과 필요에 따라 적절한 방법을 선택할 수 있습니다.

# 타이타닉 데이터를 가지고 그동안 배웠던 결측치 방법으로 모델링을 간단하게 진행해서 성능까지 평가

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.impute import KNNImputer

import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# age에 대한 결측치를 보간해서 성능을 살펴보고 개선할 수 있는 부분이 무엇인지 좀 더 고려하자!
# 고민하는 포인트를 생각해 보고 어떤 식으로 성능이 올라갈지도 좀 더 생각해 보자!
titanic.isna().sum()
```

```python
## 모델에 대한 성능을 담을 평가 데이터프레임

results = pd.DataFrame(columns = ['Method','Dataset','Accuracy','Precision','Recall','F1'])

#결측치 모두 제거 
def drop_na(df):
    return df.dropna(subset=['age']) #컬럼을 지정할 수 있음

#평균으로 대치
def mean_imputation(df):
    df_filled = df.copy()
    df_filled['age'].fillna(df_filled['age'].mean(), inplace=True)
    return df_filled

#선형보간으로 
def linear_interpolation(df):
    df_filled = df.copy()
    df_filled['age'].interpolate(method='linear',inplace=True)
    return df_filled

#KNN으로 진행
def knn_imputation(df,n_neighbors =5):
    df_filled = df.copy()
    imputer=KNNImputer(n_neighbors = n_neighbors)
    df_filled[['age']]=imputer.fit_transform(df_filled[['age']])
    return df_filled


#다양한 방법으로 결측치를 보간하고 RF 모델에 학습시켜 train,test의 생존율 예측 결과를 확인하자!

# 각 방법을 적용하여 성능 테스트
methods = {
    'Drop NA': drop_na,
    'Mean Imputation':mean_imputation,
    'Linear Interpolation':linear_interpolation,
    'Knn Imputation' : knn_imputation
}

## 타이타닉 데이터로 반복을 돌려서 방법들을 다 하나씩 돌려서 성능 테스트를 넣어야 한다.
## 데이터를 학습시키기 위해서 train, test 로 나눠야 한다.
## 인코딩하는 get_dummies 인코딩 작업

for method_name, impute_func in methods.items():
    titanic_filled = impute_func(titanic)
    
    #학습할 데이터셋으로 나누기
    X = titanic_filled.drop(['survived','deck','alone','who','adult_male','alive','embark_town'],axis = 1)
    y = titanic_filled['survived']
    
    #원핫 인코딩으로 범주형 변수를 수치로 바꾸는 것
    X = pd.get_dummies(X, drop_first=True)    
    
    #데이터셋 분할
    X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=111)
    
    model = RandomForestClassifier(random_state=111)
    model.fit(X_train, y_train)
    
    # 성능에 대해서 평가한 것들을 데이터프레임 저장해야 한다.
    
    for dataset, X_data, y_data in (['Train',X_train, y_train],('Test',X_test, y_test)):
        y_pred = model.predict(X_data)
        
        #'Accuracy','Precision','Recall','F1'
        accuracy = accuracy_score(y_data, y_pred)
        precision = precision_score(y_data, y_pred)
        recall = recall_score(y_data, y_pred)
        f1 = f1_score(y_data, y_pred)        
        
        #결과 저장
        result_now=pd.DataFrame({
            'Method':[method_name],
            'Dataset':[dataset],
            'Accuracy':[accuracy],
            'Precision':[precision],
            'Recall':[recall],
            'F1':[f1]
        })
        results = pd.concat([results, result_now], ignore_index=True)
        
print(results)
```

```python
# 시각적으로 어떤 식으로 결측치가 보간 된 걸까?

import matplotlib.pyplot as plt

fig, axes = plt.subplots(2,2, figsize=(16,12), sharex= True, sharey=True)
axes = axes.flatten() 

for ax,(method_name, impute_func) in zip (axes, methods.items()):
    titanic_filled = impute_func(titanic)
    
    ## 분포를 살펴보기
    
    sns.histplot(titanic_filled['age'], bins= 30, kde=True, ax=ax)
    ax.set_title(f'Age Distribution - {method_name}')
    ax.set_xlabel('age')
    ax.set_ylabel('Frequency')

    
#plt.subtitle('Method Com')                 
```

```python
##단순하게 결측치를 제거하는 것 보다는 어떤 도메인 로직을 통해서 결측치를 보간하는 방법도 있지 않을까?
titanic[titanic['age'].isna()]

# 요금, class, 성별이 중요한 생존에 영향을 주었다.

def fill_age_based_on_rules(df):
    # 요금을 살펴봐야 하는데 요금을 qcut 4분위로 세분화
    df['fare_bin']=pd.qcut(df['fare'], q=4, labels=['low','mid-low','mid-high','high'])
    
    # 성별, 등급, 요금의 구간별 평균 나이를 계산 할 수 있다.
    age_means=df.groupby(['sex','pclass','fare_bin'])['age'].mean()
    
    # 결측치를 보간할 때 
    def fill_age(row):
        if np.isnan(row['age']):
            try:#우리가 원하는 결측치 보간
                return age_means[row['sex'],row['pclass'],row['fare_bin']]
            except KeyError:
                return df['age'].mean() #해당 그룹이 없으면 전체 평균으로하고
        return row['age'] # 결측치 아닌 애들은 원래 값
    
    df['age'] = df.apply(fill_age, axis =1)
    df.drop(columns = ['fare_bin'], inplace=True)
    
    return df
```
