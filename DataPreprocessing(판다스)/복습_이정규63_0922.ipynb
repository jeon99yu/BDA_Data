- 데이터프레임은 원본데이터를 보존하려는 성질이 있다보니 바로 바뀌지 않는다.
- 원본 데이터를 바꾸려면 3가지를 생각해야 한다.

- 1. 원본데이터 변수명에 그대로 원본데이터를 넣으면 된다. 변수 사용해서 변수교체
- 2. 바로 문법에서 inplace=True 대치를 바로 진행하는 경우
- 3. copy() 원본데이터를 카피하여서 실제 원본데이터는 그대로 두고, 내가 분석할 데이터만 변경하여 사용한다.

- why?
     - 어느 순간 데이터를 분석하다보면 뭐가 원본이고 무엇을 하고 있는지? 가끔 혼선이 올 경우
     - A로 전처리 했는데- B로 전처리 해야 한다. 원본 없다? 귀찮은 일
     - 이런 여러 예외 케이스를 대응하기 위해서는 원본 데이터는 그대로 두는 게 처음부터 습관! 좋다!

### 데이터 전처리, 추출 등에 필요한 기초 문법
     - 행과 열의 기준 외 특정 데이터 추출 조건에 따른 데이터를 추출 작업!
     - loc,iloc 행,열 좌표 위치
     - 성별이 여자인 경우 또는 성별이 남자인 경우
- query()- 데이터 추출
- df.loc/iloc[불린참값] - 데이터 추출
- df[] -열 기준 추출
- sort_values() 정렬
- group by() 그룹바이
- assign() 변수 추가 할 때 사용하는 함수
- merge() -  병합
- concat() - 병합
- apply() -전처리 전체 적용

```python
#성별이 남성인 경우
#query(특정 요구조건 적으면 된다.)
## 문자열 데이터는 같은지 여부로 데이터 추출 가능
tt.query("sex == 'male'")
```

```python
## 수치형 데이터는?
## 대소비교 > < >= =<
tt.query('fare >50')
```

```python
## 다중 조건인 경우?
## 요금은 50 이상에 나이는 20이상 다중조건
## and, or조건 값이 달라집니다.
tt.query('fare>50 & age <20')
```

```python
#and 조건, or조건
tt.query('fare>50 | age <20')
```

```python
tt.query("embark_town in ['Cherbourg','Southampton']")
```
